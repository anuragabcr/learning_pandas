```python
columns_to_drop = [
    'id', 'reviews per month', 'review rate number', 'calculated host listings count', 'availability 365', 'house_rules', 'license'
]
columns_to_keep = [
    'NAME', 'host id', 'host_identity_verified', 'host name', 'neighbourhood group', 'neighbourhood', 'lat', 'long', 'country', 'country code', 'instant_bookable', 'cancellation_policy', 'room type', 'Construction year', 'price', 'service fee', 'minimum nights', 'number of reviews', 'last review'
]
```

### 1st Method, Filtering Data

```python
df = data[columns_to_keep]
df.head()
```

```python
df.shape
```

### 2nd Method, Dropping Columns

```python
data.drop(columns=['license'])
```

```python
data.drop(columns=columns_to_drop, inplace=True)
```

```python
data.shape
```

### 3rd Method, Rename columns

```python
data.rename(columns={"NAME": "Name", "host id": "host_Id"})
```

```python
new_columns = []
for column in data.columns:
    new_columns.append(column.replace(" ", "_").lower())
```

```python
new_data = data.copy()
new_data.columns = new_columns
```

```python
new_data.columns
```

### 4th Method, Drop Duplicate

```python
data.duplicated().sum()
```

```python
data.duplicated().value_counts()
```

```python
data = data.drop_duplicates()
```

```python
data.duplicated().sum()
```

### 5th Method, Handle null

```python
data.isna().sum()
```

```python
data.drop(columns=['last review'], inplace=True)
```

```python
data.isna().sum()
```

```python
data.dropna(inplace=True)
```

```python
data.isna().sum()
```

```python
data.shape
```

6th Method, Handle single column

```python
data["host_identity_verified"].str.upper()
```

```python
data["host_identity_verified"] = data["host_identity_verified"].str.upper()
```

```python
data.head()
```

```python
data["instant_bookable"] = data["instant_bookable"].apply(lambda x: 1 if x else 0)
```

```python
data.head()
```

```python
data.reset_index(drop=True, inplace=True)
```

```python
data.head()
```

```python
data["price"] = data["price"].str.replace("$", "").str.replace(",", "").astype(float)
```
