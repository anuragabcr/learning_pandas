```python
bios['height_category'] = pd.cut(bios['height_cm'], bins=[0, 180, 200, 250], labels=['Short', 'Average', 'Tall'])
```

```python
bios['height_cat'] = bios['height_cm'].apply(lambda x: 'Tall' if x > 200 else ('Average' if x > 180 else 'Short'))
```

```python
bios.head()
```

```python
def categorize_athelete(row):
    if row['height_cm'] < 175 and row['weight_kg'] < 70:
        return 'Lightweight'
    elif row['height_cm'] <= 200 and row['weight_kg'] <= 100:
        return 'Middleweight'
    else:
        return 'Heavyweight'

bios['catehory'] = bios.apply(categorize_athelete, axis=1)
```
